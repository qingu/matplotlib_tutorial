## Pyplot模块

matplotlib的pyplot模块提供了和MATLAB类似的绘图API，使得我们能够快速的绘制二维图形。每个pyplot函数对一张图表（figure）做一些改变，例如创建图表、创建图表绘图区域、在绘图区域绘制图线、使用标签装饰图表等。

matplotlib实际上是一套面向对象的绘图库，所绘制的图表中的每个绘图元素，例如线条、文字、刻度等在内存中都有一个对象与之对应。为了将面向对象的绘图库封装成只使用函数的调用接口，pyplot模块的内部保存了当前图表以及当前绘图区域等信息。绘图函数指向当前axes（matplotlib中的axes指图表的坐标轴部分，而不是数学意义上的多个坐标轴）。

pyplot API相对于面向对象的API灵活性差些。在需要精细绘制时我们可以使用对应的Axes对象的方法代替pyplot函数。

下面通过一个简单例子将图表绘制的要素通过pyplot函数展示出来

````python
import numpy as np
import matplotlib.pyplot as plt

#准备数据
x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.cos(x**2)

#开始绘制图表
#figure()创建一个Figure对象，可以设置图表大小（单位：英寸或dpi像素）
plt.figure(figsize=(8,4))

#创建Figure对象后，调用plot()在当前Figure对象的Axes对象上绘图
#plot()返回Line2D对象
plt.plot(x,y,label=r"$sin(x)$",color="red",linewidth=2)
plt.plot(x,z,"b--",label=r"$cos(x^2)$")

#设置当前Axes对象的各个属性
#xlabel()设置X、Y坐标轴标题文字
plt.xlabel("Times(s)")
plt.ylabel("Volt")

#title()设置当前绘图区域的标题
plt.title("Pyplot First Example")

#xlim()、ylim()设置X、Y轴的显示范围，坐标范围
plt.ylim(-1.2,1.2)

#legend()显示图例
plt.legend()

#show()显示出绘图窗口，当然也可以通过savefig()将当前Figure对象保存为图像文件
plt.show
````

上面的例子我们是直接调用函数的方式进行绘制，实际上许多函数都会返回一个对象可供进一步操作。我们是通过pyplot内部对象隐藏了该信息。

pyplot.plot()函数可以接受列表类型数据，更常见的是numpy数组。实际上所有序列在内部转换成了numpy数组。下面我们使用列表作为例子看看。

````python
import matplotlib.pyplot as plt

#accept list type argument
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
         
#axis([xmin, xmax, ymin, ymax])
plt.axis([0, 6, 0, 20])
plt.show()
````

plot()可以接受任意多个参数。

````python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')

plt.show()
````

### 关键字字符串绘图

常常我们的数据是类似字典类型的，可以通过字符串访问，例如pandas.DataFrame对象。

Matplotlib允许通过`data`关键字参数提供这样的参数。

````python
#字典类型的数据
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

#data关键子参数指定数据类型，然后数据参数部分通过字符串指定访问特定数据
#c -- color, s -- scalar
plt.scatter('a', 'b', c='c', s='d', data=data)

plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

### 分类变量

````python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````
